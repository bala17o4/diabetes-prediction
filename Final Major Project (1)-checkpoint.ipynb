<h3>Data Analyzing</h3>
<p>We can see that info gives that there are no null values or no nan values; but if we analyze the dataset using excel we can see that
some variables like Glucose, Blood Pressure, Skin Thickness, Insulin, BMI have values as 0, which is not possible in real world, so for us it is still
null values or nan values, and we must do some preprocessing to deal with them</p>

```python
col_pb = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
```

```python
for col in col_pb:
    df[col].replace(to_replace = 0, value = np.nan, inplace = True)
```

```python
df.head()
```

```python
df1 = df.copy()
```

```python
df1.isnull().sum()
```

<h3> Preprocessing Steps</h3>
<p> From now we will use some different preprocessing steps</p>
<p> As Glucose, BMI, Blood Pressure have less number of NaN values we can drop these rows in Glucose, BMI, Blood Pressure</p>                

```python
df1.dropna(axis = 0, subset=['Glucose', 'BMI', 'BloodPressure'], inplace = True)
```

```python
df1['Outcome'].value_counts()
```

<p> We can also see that dataset is not uniform, It has more number of non-diabetic patients than diabetic</p>

### Filling the missing values with mean 
<p> The columns which have large number of null values, their NaN values will be replaced by mean of these variables</p>

```python
df1['SkinThickness'].replace(to_replace=np.nan, value = np.mean(df1['SkinThickness']), inplace = True)
```

```python
df1['Insulin'].replace(to_replace=np.nan, value = np.mean(df1['Insulin']), inplace = True)
```

```python
df1.isnull().sum()
```

```python
df1.shape
```

```python
plt.figure(figsize=(30, 10))
sns.set_theme(style='whitegrid')
i = 1
for col in df1.columns:
    if col=='Outcome': 
        continue
    plt.subplot(2, 4, i)
    sns.boxplot(x = df1[col], hue='Outcome', data=df1)
    i = i+1
```

```python
df1.describe()
```

#### Removing Outliers

<p>Searching on Google and with self-assumptions; we remove the outliers from data, as we can see and search that Diastolic Blood Pressure below 40 is very rare, so we can remove values of Blood Pressure below 40, same can be done with Skin Thickness, BMI and Insulin </p>

```python
df1 = df1[~(df1['Pregnancies']>13)]
```

```python
df1 = df1[~(df1['BloodPressure']<40)]
```

```python
df1 = df1[~(df1['SkinThickness']==99)]
df1 = df1[~(df1['BMI']==67.1)]
df1 = df1[~(df1['Insulin']>500)]
```

```python
df1.shape
```

```python
df1.describe()
```

```python
X = df1.drop(['Outcome'], axis = 1)
y = df1['Outcome']
```

#### Splitting the dataset with X(Features) and Y(Outcomes) into training and testing Data

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

<p style="color:red">Using <u>Standard Scaler</u> for scaling input, for normalizing the feature variables</p>
<p><b>(X(i) - X(mean))/STD(X)</b></p>

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train_norm = scaler.transform(X_train)
X_test_norm = scaler.transform(X_test)
```

<h4>Logistic Regression</h4>
<p><b>Using Logistic Regression to train the model and testing on X_test(Test Dataset) to make predictions and judging the model using various classification metrics like accuracy score, confusion matrix, and f1_score</b></p>
<p><b> Logistics Regression is a Classification matrix and it uses <i>Sigmoid Function</i>, to make predictions</b></p> 

```python
from sklearn.linear_model import LogisticRegression as LR
```

```python
log = LR()
log.fit(X_train_norm, y_train)
pred = log.predict(X_test_norm)
print("LOGISTIC REGRESSION")
print("Accuracy Of Logistic Regression: ",accuracy_score(y_test, pred))
print(confusion_matrix(y_test, pred))
print(f1_score(y_test, pred))
```

<h4>K Nearest Neighbors</h4>
<p><b>Using KNN to train the model and testing on X_test(Test Dataset) to make predictions and judging the model using various classification metrics like accuracy score, confusion matrix, and f1_score</b></p>
<p><b> K Nearest Neighbors is a Classification algorithm and it uses <u>Euclidean Distance or some other distance metric to consider k nearest neighbors to itself and then choose the class with majority of occurences among them</u>, to make predictions</b></p> 

```python
knn = KNeighborsClassifier()
knn.fit(X_train_norm, y_train)
pred = knn.predict(X_test_norm)
print("KNearestNeighbors")
print("Accuracy of KNearestNeighbors is: ",accuracy_score(y_test, pred))
print(confusion_matrix(y_test, pred))
print(f1_score(y_test, pred))
```

### Now Trying a Different method for Nan Values

<h4>Dealing With missing Values</h4>
<p style="color: red">We will first remove all the null values of the respective variables and then compute the median variable seperately for non-diabetic and diabetic patients</p>

```python
##This Function does for each variable the above mentioned process
def median_target(var):   
    temp = df[df[var].notnull()]
    temp = temp[[var, 'Outcome']].groupby(['Outcome'])[[var]].median().reset_index()
    return temp
```

```python
columns = df3.columns
columns = columns.drop("Outcome")
for i in columns:
    median_target(i)
    df3.loc[(df3['Outcome'] == 0 ) & (df3[i].isnull()), i] = median_target(i)[i][0]
    df3.loc[(df3['Outcome'] == 1 ) & (df3[i].isnull()), i] = median_target(i)[i][1]
```
